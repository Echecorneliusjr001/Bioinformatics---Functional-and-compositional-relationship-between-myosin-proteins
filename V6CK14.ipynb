# Using the for loop to iterate through in order to get the excel file that contains number of GO Terms , start and without myos column


```python
# peptide that I am interested in (non-myosin one now)
peptides_of_interest = ["V6CK14_CAEEL"] 

# List to store final results
output_rows = []

# Keywords for myosins to exclude
myosin_keywords = "myo-3|myo-1|unc-54|myo-2"

for peptide in peptides_of_interest:
    print(f"Processing: {peptide}")

    # Filter entries matching the peptide
    matching_entries = df_pkl[df_pkl["Entry name"] == peptide]

    for _, row in matching_entries.iterrows():
        result_row = {
            "Start": row["Start"],
            "Kclust": row["Kclust"]
        }

        # Get all entries with the same Kclust
        kclust_filter = df_pkl["Kclust"] == row["Kclust"]
        kclust_df = df_pkl[kclust_filter]

        # Filter out rows containing myosin-related genes
        no_generic_filter = ~kclust_df["Gene Names (primary)"].str.contains("somestuff", na=False)
        no_myosin_filter = ~kclust_df["Gene Names (primary)"].str.contains(myosin_keywords, na=False)

        if no_myosin_filter.sum() == no_generic_filter.sum():
            print(f"Kclust {row['Kclust']} does not have myosins.")
            result_row["Othermyos"] = False
        else:
            result_row["Othermyos"] = True

        # Save the filtered Kclust group to Excel
        output_filename = f"pixel_{peptide}_{row['Start']}.xlsx"
        kclust_df.to_excel(output_filename, index=False)
        print(f"Saved to {output_filename}")

        output_rows.append(result_row)

# Save summary results
summary_df = pd.DataFrame(output_rows)
summary_df.to_excel("resdf3.xlsx", index=False)
print("Summary saved to resdf3.xlsx")
```

```python
from gprofiler import GProfiler
import pandas as pd
import numpy as np

# Loaded the summary results
summary_df = pd.read_excel("resdf3.xlsx")

# Initialized GProfiler and results list
gp = GProfiler(return_dataframe=True)
GOs = []

# Filtered clusters without myosins
no_myosin_clusters = summary_df[summary_df["Othermyos"] == False]["Kclust"].unique()

# Run enrichment analysis for each Kclust
for kclust_id in no_myosin_clusters:
    print(f"Running enrichment for Kclust {kclust_id}...")

    # Getting genes in the current Kclust
    subdf = df_pkl[df_pkl["Kclust"] == kclust_id]
    gene_list = list(subdf["Gene Names (primary)"].dropna().unique())

    output_go_file = f"cluster_{kclust_id}_GOs.xlsx"

    # If no valid gene names, write an empty dataframe with headers
    if not gene_list:
        print(f"No valid gene names for Kclust {kclust_id}. Writing empty file.")
        pd.DataFrame(columns=["Kclust", "qgenelist"]).to_excel(output_go_file, index=False)
        continue

    # Run GO enrichment
    results = gp.profile(organism='celegans', query=gene_list, no_evidences=False)

    # Filtered out unwanted sources
    results = results[~results["source"].isin(["TF", "MIRNA", "HP"])]

    # Added metadata
    results["Kclust"] = kclust_id
    results["qgenelist"] = str(gene_list)

    # Saved individual result, even if was empty
    results.to_excel(output_go_file, index=False)
    print(f"Saved enrichment results to {output_go_file}")

    # Added to list only if enrichment results exist
    if not results.empty:
        GOs.append(results)

# Saved combined results (only non-empty ones)
if GOs:
    pd.concat(GOs).to_excel("cluster_V6CK14_GOs.xlsx", index=False)
    print("Saved combined enrichment results to cluster_V6CK14_GOs.xlsx")
else:
    print("No enrichment results to combine.")

```

```python
# Used for loop to iterate over the resdf3 file and count the number of GO terms in each cluster file
import numpy as np
lengo = []
for i, row in resdf3.iterrows():
    Kclust = row["Kclust"]
    filename = f"cluster_{Kclust}_GOs.xlsx"
    
    # Assign 0 if file is not found, to avoid limiting computation
    try:
        godf = pd.read_excel(filename)
        lengo.append(len(godf))
    except FileNotFoundError:
        print(f"File {filename} not found. Assigning 0 GO terms.")
        lengo.append(0)

resdf3["Number of GO term"] = lengo
resdf3.to_excel("resdf4.xlsx", index=False)
print("Saved resdf4.xlsx with GO term counts.")
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
columns_to_extract = [
    'CA-Gly', 'Pro-MC', 'CB', 'CB-Pro', 'Carboxyl',
    'Amide', 'His', 'Trp', 'Phe-Tyr', 'OH-Tyr', 'CG',
    'CD', 'CE', 'OH', 'SH', 'S', 'NH3', 'Arg', 'MC'
]

def standardize_max(reference_data, data):
    max_values = np.max(reference_data, axis=1)
    min_values = np.min(reference_data, axis=1)
    return [
        (col - min_val) / (max_val - min_val) if max_val != min_val else col * 0
        for col, max_val, min_val in zip(data, max_values, min_values)
    ]
```

```python
df_20_000 = pd.read_pickle('mergedbf_unusual_marked_with_20000cluster_label (1).pkl')
df_20_000 = df_20_000.rename(columns={'Kclust':'cluster'})
```

```python
df_clust = df_20_000
clust = df_clust[df_clust.cluster == 560] # c.elegans 280 start(AA position) THE HIGHEST NUMBER OF GO TERMS of an amino acid position in highly enrinched region
full_proteome_data = [df_clust[col] for col in columns_to_extract]
clust_data = standardize_max(full_proteome_data, [clust[col] for col in columns_to_extract])
normalized_full = standardize_max(full_proteome_data, full_proteome_data)

boxplot_data = []
positions = []
xtick_labels = []
offset = 0.2

for i, (feature, clust_col, full_col) in enumerate(zip(columns_to_extract, clust_data, normalized_full)):
    base_pos = i
    positions.append(base_pos - offset)
    boxplot_data.append(clust_col)
    positions.append(base_pos + offset)
    boxplot_data.append(full_col)
    xtick_labels.append(feature)

fig, ax = plt.subplots(figsize=(12, 8))

colors = ['steelblue', 'gray'] * len(columns_to_extract)
bp = ax.boxplot(
    boxplot_data,
    positions=positions,
    widths=0.35,
    patch_artist=True,
    whis=(0, 100),
    boxprops=dict(linewidth=1.2),
    medianprops=dict(linewidth=1.5, color='red')
)


for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

whiskers = [True,True,False,False] * len(columns_to_extract)

for patch, whisker in zip(bp['whiskers'], whiskers):
    
    if whisker==False:
        patch.remove()
    else:
        patch.set_color('steelblue')
        patch.set_lw(2.0)

for patch, whisker in zip(bp['caps'], whiskers):
    if whisker==False:
        patch.remove()
    else:
        patch.set_color('steelblue')
        patch.set_lw(2.0)

ax.set_xticks(range(len(columns_to_extract)))
ax.set_xticklabels(columns_to_extract, fontsize=20, rotation=90)
ax.set_yticks(np.linspace(0, 1, 11))
ax.set_ylabel("Normalized feature count", fontsize=25)
ax.set_xlabel("Functional groups or atoms", fontsize=25)
ax.tick_params(axis='y', labelsize=18)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
for spine in ax.spines.values():
    spine.set_linewidth(2)
ax.tick_params(width=2)

legend_elements = [
    Patch(facecolor='steelblue', label='Cluster'),
    Patch(facecolor='gray', label='Proteome')
]
ax.legend(handles=legend_elements, loc='upper center', 
          #bbox_to_anchor=(0.5, 1.1),
          ncol=2, fontsize=20)

plt.tight_layout(pad=2)
plt.title("Standardized Feature Boxplots for pixel_V6CK14_CAEEL_280 ", fontsize=22, fontweight='bold')
filename='feat_clust_560_pos_280'
fig.savefig(filename+'.png')
```
