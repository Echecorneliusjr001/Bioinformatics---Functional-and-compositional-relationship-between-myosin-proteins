```python
# Extracting columns
columns_to_extract = [
    'CA-Gly', 'Pro-MC', 'CB', 'CB-Pro', 'Carboxyl',
    'Amide', 'His', 'Trp', 'Phe-Tyr', 'OH-Tyr', 'CG',
    'CD', 'CE', 'OH', 'SH', 'S', 'NH3', 'Arg', 'MC'
]

# define standardization function
def standardize_max(reference_data, data):
    max_values = np.max(reference_data, axis=1)
    min_values = np.min(reference_data, axis=1)
    return [
        (col - min_val) / (max_val - min_val) if max_val != min_val else col * 0
        for col, max_val, min_val in zip(data, max_values, min_values)
    ]
    
   
    
```

```python
df_clust = df_20_000
clust = df_clust[df_clust.cluster == 2640] # c.elegans 1780 start(AA position) Reading Excel file with high GO terms in relation to rod-like
# tail region for feature extraction with 82 GO terms

full_proteome_data = [df_clust[col] for col in columns_to_extract]
clust_data = standardize_max(full_proteome_data, [clust[col] for col in columns_to_extract])
normalized_full = standardize_max(full_proteome_data, full_proteome_data)

boxplot_data = []
positions = []
xtick_labels = []
offset = 0.2

for i, (feature, clust_col, full_col) in enumerate(zip(columns_to_extract, clust_data, normalized_full)):
    base_pos = i
    positions.append(base_pos - offset)
    boxplot_data.append(clust_col)
    positions.append(base_pos + offset)
    boxplot_data.append(full_col)
    xtick_labels.append(feature)

fig, ax = plt.subplots(figsize=(12, 8))

colors = ['steelblue', 'gray'] * len(columns_to_extract)
bp = ax.boxplot(
    boxplot_data,
    positions=positions,
    widths=0.35,
    patch_artist=True,
    whis=(0, 100),
    boxprops=dict(linewidth=1.2),
    medianprops=dict(linewidth=1.5, color='red')
)


for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

whiskers = [True,True,False,False] * len(columns_to_extract)

for patch, whisker in zip(bp['whiskers'], whiskers):
    
    if whisker==False:
        patch.remove()
    else:
        patch.set_color('steelblue')
        patch.set_lw(2.0)

for patch, whisker in zip(bp['caps'], whiskers):
    if whisker==False:
        patch.remove()
    else:
        patch.set_color('steelblue')
        patch.set_lw(2.0)

ax.set_xticks(range(len(columns_to_extract)))
ax.set_xticklabels(columns_to_extract, fontsize=20, rotation=90)
ax.set_yticks(np.linspace(0, 1, 11))
ax.set_ylabel("Normalized feature count", fontsize=25)
ax.set_xlabel("Functional groups or atoms", fontsize=25)
ax.tick_params(axis='y', labelsize=18)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
for spine in ax.spines.values():
    spine.set_linewidth(2)
ax.tick_params(width=2)

legend_elements = [
    Patch(facecolor='steelblue', label='Cluster'),
    Patch(facecolor='gray', label='Proteome')
]
ax.legend(handles=legend_elements, loc='upper center', 
          #bbox_to_anchor=(0.5, 1.1),
          ncol=2, fontsize=20)

plt.tight_layout(pad=2)
plt.title("Standardized Feature Boxplots for pixel_MYO2_CAEEL_1780 ", fontsize=22, fontweight='bold')
filename='feat_clust_2640_pos_1780'
fig.savefig(filename+'.png')
```
